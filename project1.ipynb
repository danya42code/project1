B: Обработка данных

```python
df.isnull().sum()
```

```python
df.dropna(subset='page_count', inplace=True)
```

```python
df['title'].fillna(" ", inplace=True)
df['subtitle'].fillna(" ", inplace=True)
df['authors'].fillna("Anonim", inplace=True)
df['publisher'].fillna("Internet", inplace=True)
df['published_date'].fillna("2000-01-01", inplace=True)
df['published_date'] = np.where(df['published_date'].str.len()==4, df['published_date']+'-01-01', df['published_date'])
df['published_date'] = np.where(df['published_date'].str.len()==7, df['published_date']+'-01', df['published_date'])
df['description'].fillna("No description", inplace=True)
df['categories'].fillna("Other", inplace = True)
df['average_rating'].fillna(df['average_rating'].mean(), inplace=True)
df['isbn_13'].fillna(0, inplace=True)
df['isbn_10'].fillna(0, inplace=True)
df['list_price'] = np.where(df['buyable'], df['list_price'].mean(), 0)
df['currency'].fillna('USD', inplace=True)
df['thumbnail'].fillna("No link", inplace=True)
```

```python
df.isnull().sum()
```

```python
df['published_date'] = df['published_date'].str[:4]
df['published_date'] = df['published_date'].str.replace('?', '5')
df['published_date'].astype(int)
```
